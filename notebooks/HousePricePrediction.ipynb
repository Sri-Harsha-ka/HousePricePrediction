```python
y = df['price']
X = df.drop(['price'] , axis=1)
```

```python
from sklearn.model_selection import train_test_split

X_train , X_test , y_train,y_test = train_test_split(X,y ,test_size=0.1 , random_state=42)
X_train.shape , y_train.shape , X_test.shape
```

```python
X_train.head()
```

```python
X_train_rf = X_train.copy()
X_test_rf = X_test.copy()
```

```python
# This is ok but Lets do better 
# pd.get_dummies(df , columns=['zipcode'] , drop_first=True)

train_df = X_train.copy()
train_df['price'] = y_train

# Get the Median for each place
zip_map = train_df.groupby('zipcode')['price'].median()

# replace them in the each data
X_train['zipcode_encoded'] = X_train['zipcode'].map(zip_map)
X_test['zipcode_encoded'] = X_test['zipcode'].map(zip_map)
X_train.drop('zipcode' , axis=1 ,inplace=True)
X_test.drop('zipcode' , axis=1 ,inplace=True)
```

```python
from sklearn.preprocessing import StandardScaler

sc = StandardScaler()
X_train_scaled = sc.fit_transform(X_train)
X_test_scaled = sc.transform(X_test)
```

```python
X_train.head()
```

```python
# First Model or Base Model 
from sklearn.linear_model import LinearRegression

lr_model = LinearRegression()
lr_model.fit(X_train_scaled , y_train)
```

```python
# Predictions of the model

lr_preds = lr_model.predict(X_test_scaled)
```

```python
from sklearn.metrics import r2_score , mean_absolute_error , mean_squared_error

print(r2_score(lr_preds, y_test))
print(mean_absolute_error(lr_preds , y_test))
print(mean_squared_error(lr_preds , y_test))
```

```python
X_train_599 = X_train.loc[[599]]
X_scaled_599 = sc.transform(X_train_599)
prediction = lr_model.predict(X_scaled_599)
prediction
```

```python
# Ridge Model 

from sklearn.linear_model import Ridge
ridge_model = Ridge(alpha=0.01)
ridge_model.fit(X_train_scaled,y_train)
```

```python
ridge_preds = ridge_model.predict(X_test_scaled)

print(r2_score(ridge_preds, y_test))
print(mean_absolute_error(ridge_preds , y_test))
print(mean_squared_error(ridge_preds , y_test))
```

```python
# Random Forest 

from sklearn.ensemble import RandomForestRegressor

rf_model = RandomForestRegressor(
    n_estimators=500,
    max_depth=None,
    min_samples_split=2,
    min_samples_leaf=1,
    max_features='sqrt',
    random_state=42,
    n_jobs=-1
)

rf_model.fit(X_train_rf,y_train)
```

```python
rf_preds = rf_model.predict(X_test_rf)

print(r2_score(rf_preds, y_test))
print(mean_absolute_error(rf_preds , y_test))
print(mean_squared_error(rf_preds , y_test))
```

```python
import joblib
joblib.dump(lr_model,"LR_Model.pkl")
joblib.dump(sc , "Scalar.pkl")
joblib.dump(X_train.columns.tolist(), "Columns.pkl")
```
